```python
# --- Agglomerative Hierarchical Clustering on Vehicle Dataset ---

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.preprocessing import StandardScaler

# 1. Create Vehicle Dataset
data = {
    'Vehicle': ['Sedan', 'SUV', 'Sports Car', 'Truck', 'Hatchback', 'Minivan', 'Coupe'],
    'Horsepower': [130, 200, 350, 250, 110, 180, 280],
    'Weight': [1500, 2000, 1400, 3000, 1200, 1800, 1600],
    'Top Speed': [180, 190, 300, 160, 170, 175, 250]
}

df = pd.DataFrame(data)
X = df[['Horsepower', 'Weight', 'Top Speed']]

# 2. Standardize Features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. Linkage Methods
linkage_methods = ['single', 'complete', 'average']

# 4. Plot Dendrograms
plt.figure(figsize=(18, 5))
for i, method in enumerate(linkage_methods):
    plt.subplot(1, 3, i+1)
    Z = linkage(X_scaled, method=method)
    dendrogram(Z, labels=df['Vehicle'].values, leaf_rotation=45)
    plt.title(f'Agglomerative Clustering ({method} linkage)')
    plt.xlabel('Vehicle')
    plt.ylabel('Distance')

plt.tight_layout()
plt.show()
```
